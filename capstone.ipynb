HYPOTHESIS TESTING

One sample t test

Considering first 3950 selling prices of the data set as sample (half of dataset)

```python
mean_selling_price = car_dataset['selling_price'].mean()
mean_selling_price
```

```python
from scipy.stats import ttest_1samp
tstat, pval = ttest_1samp(car_dataset["selling_price"][1:3950], mean_selling_price)
pval
```

As the p value is greater than 0.05 the null hypothesis is failed to be rejected and the mean value of sample is same as population

Considering first 3950 km driven of the data set as sample (half of dataset)

```python
mean_km_driven = car_dataset['km_driven'].mean()
mean_km_driven
```

```python
from scipy.stats import ttest_1samp
tstat, pval = ttest_1samp(car_dataset["km_driven"][1:3950], mean_km_driven)
pval
```

As the p value is greater than 0.05 the null hypothesis is failed to be rejected and the mean value of sample is same as population

Two sample t-test

Considering 3950 selling prices of the data set as two sample

```python
mean_1 = car_dataset['selling_price'][1:3950].mean()
mean_1
```

```python
mean_2 =car_dataset['selling_price'][3950:].mean()
mean_2
```

```python
from scipy.stats import ttest_ind
tstat, pval = ttest_ind(car_dataset['selling_price'][1:3950],car_dataset['selling_price'][3950:])
print("p-value",pval)
```

As p-val is greater than 0.05 we fail to reject the null hypothesis and samples are related with each other

Considering 3950 km driven of the data set as two samples

```python
mean_km_1 = car_dataset['km_driven'][1:3950].mean()
mean_km_1
```

```python
mean_km_2 = car_dataset['km_driven'][3950:].mean()
mean_km_2
```

```python
from scipy.stats import ttest_ind
tstat, pval = ttest_ind(car_dataset['km_driven'][1:3950],car_dataset['km_driven'][3950:])
print("p-value",pval)
```

As p-val is greater than 0.05 we fail to reject the null hypothesis and samples are related with each other

One Way ANOVA

```python
import researchpy as rp
rp.summary_cont(car_dataset['selling_price'].groupby(car_dataset['City']))
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

model = ols('selling_price ~ C(City)', data=car_dataset).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

As p value(PR) less than 0.05 the null hypothesis is rejected and there is significant difference between different city selling price

TWO WAY ANOVA

```python
formula = 'selling_price ~ C(City) + C(Region) + C(City):C(Region)'
model = ols(formula, car_dataset).fit()
aov_table = sm.stats.anova_lm(model, typ=2)
aov_table
```

Salling price varies in case of city and the combination of city and region, but there is no much variance in case of region. As,the null hypothesis is rejected in case of city and combination of city and region. It is accepetd n case of Region.

Chi-Square Test

```python
df=car_dataset[car_dataset['sold']=='Y'].groupby('City').agg({'selling_price':np.sum,'sold':np.size})
df
```

```python
from scipy import stats
chi_stat,pval,dof,expected_values = stats.chi2_contingency(df)
#stats.chi2_contingency(df)
pval
```

```python
from scipy.stats import chi2
alpha = 0.05
critical_value = chi2.ppf(q=1-alpha, df=dof)
print('critical_value:', critical_value)
```

```python
chi_stat>=critical_value
```

Reject Null hypothesis as the chi_stat value >= critical value. There is a relationship between selling price,sold based on city
